```python
import onnxruntime as ort
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the ONNX model
onnx_model_path = 'C:/Users/omhas/OneDrive/Documents/YOLO Project/2_Prediction_Model/train/weights/best.onnx'  # Replace with the path to your best.onnx file
session = ort.InferenceSession(onnx_model_path)

# Get input and output details
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name
input_shape = session.get_inputs()[0].shape  # Example: [1, 3, 640, 640]

# Load the input image
image_path = 'C:/Users/omhas/OneDrive/Documents/YOLO Project/2_Prediction_Model/street_image.jpg'
image = cv2.imread(image_path)

# Load class names from data.yaml
class_names = [
    'person', 'chair', 'car', 'dog', 'bottle', 'cat', 'bird', 'pottedplant', 'sheep', 'boat', 
    'aeroplane', 'tvmonitor', 'sofa', 'bicycle', 'horse', 'motorbike', 'diningtable', 'cow', 'train', 'bus'
]

# Preprocess the image
def preprocess(image, input_shape):
    # Resize the image to match the model's input shape
    resized_image = cv2.resize(image, (input_shape[3], input_shape[2]))  # Width and height
    # Normalize the image (scale to 0-1 and apply mean/std if required)
    blob = resized_image.astype(np.float32) / 255.0
    # Transpose to [batch, channels, height, width]
    blob = np.transpose(blob, (2, 0, 1))  # HWC to CHW
    blob = np.expand_dims(blob, axis=0)  # Add batch dimension
    return blob

# Preprocess the input image
blob = preprocess(image, input_shape)

# Perform inference
outputs = session.run([output_name], {input_name: blob})[0]

# Postprocess the output
def postprocess(outputs, image, class_names, conf_threshold=0.5, nms_threshold=0.4):
    height, width = image.shape[:2]
    boxes = []
    confidences = []
    class_ids = []

    # Iterate over the output detections
    for detection in outputs[0].T:  # Transpose to iterate over detections
        # Extract bounding box coordinates (center_x, center_y, width, height)
        center_x, center_y, w, h = detection[:4]

        # Extract objectness score and normalize it using sigmoid
        objectness = detection[4]
        objectness = 1.0 / (1.0 + np.exp(-objectness))  # Apply sigmoid

        # Extract class probabilities and normalize them using softmax
        class_probs = detection[5:24]  # Adjust this if the number of classes is different
        class_probs = np.exp(class_probs) / np.sum(np.exp(class_probs))  # Apply softmax

        # Calculate confidence score
        class_id = np.argmax(class_probs)
        confidence = objectness * class_probs[class_id]

        if confidence > conf_threshold:
            # Convert to x1, y1, x2, y2 format
            x1 = int((center_x - w / 2) * width)
            y1 = int((center_y - h / 2) * height)
            x2 = int((center_x + w / 2) * width)
            y2 = int((center_y + h / 2) * height)

            boxes.append([x1, y1, x2, y2])
            confidences.append(float(confidence))
            class_ids.append(int(class_id))

    # Perform non-maximum suppression to filter overlapping boxes
    indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)

    # Draw bounding boxes and labels on the image
    if len(indices) > 0:
        for i in indices.flatten():
            x1, y1, x2, y2 = boxes[i]
            class_id = class_ids[i]
            confidence = confidences[i]

            # Draw bounding box
            color = (0, 255, 0)  # Green color for bounding box
            cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)

            # Create label with class name and confidence
            label = f"{class_names[class_id]}: {confidence:.2f}"

            # Put label on the image
            cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return image

# Postprocess the output and draw bounding boxes
output_image = postprocess(outputs, image, class_names, conf_threshold=0, nms_threshold=10)  # Lower the threshold

# Convert BGR image to RGB for displaying with matplotlib
output_image_rgb = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)

# Display the output image
plt.figure(figsize=(10, 10))
plt.imshow(output_image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```
